# Old Testament Word Clouds and Frequency Distributions

```python
import os, nltk, wordcloud, re
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk import FreqDist, word_tokenize
```

```python
%matplotlib inline
```

**Download text copies of the Old Testament for analysis**

```python
study_dir = './OldTestament/'
books = os.listdir('OldTestament/')
books = [i for i in books if i.startswith('1')]
```

```python
books
```

**Read the books in as strings**

```python
with open((study_dir + books[2]), 'r' ) as f:
    Genesis = f.read()
with open((study_dir + books[1]), 'r' ) as f:
    Exodus = f.read()
with open((study_dir + books[0]), 'r' ) as f:
    Leviticus = f.read()
with open((study_dir + books[3]), 'r' ) as f:
    Numbers = f.read()
with open((study_dir + books[4]), 'r' ) as f:
    Deuteronomy = f.read()
```

```python
Genesis[0:57]
```

**We will need to tokenize and strip punctuation from the strigs for the Frequency Charts**  
**Additionally with the Neuro Linguistic toolkit we can strip the stopwords**  
**you may have to use nltk.download('punct') in order to use the english**  
**punctuation inthe tokenizer**  

```python
punc = re.compile('\w+')
Genesis_Frequency = FreqDist([word for word in word_tokenize(Genesis) if word not in stopwords.words() and punc.match(word)])
```

```python
def Book_Frequency(book):
    punc = re.compile('\w+')
    Book_Frequency = FreqDist([word for word in word_tokenize(book) if word not in stopwords.words() and punc.match(word)])
    return Book_Frequency
```

** Now we can use the Book_Frequency method to Render the rest of the Frequencies **

```python
Exodus_Frequency = Book_Frequency(Exodus)
Leviticus_Frequency = Book_Frequency(Leviticus)
Number_Frequency = Book_Frequency(Numbers)
Deuteronomy_Frequency = Book_Frequency(Deuteronomy)
```

```python
Genesis_Frequency.plot(18)
```

```python
Exodus_Frequency.plot(20)
```

```python
Leviticus_Frequency.plot(24)
```

```python
Number_Frequency.plot(20)
```

```python
Deuteronomy_Frequency.plot(20)
```
